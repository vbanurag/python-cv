```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img, cmap='gray'):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')
```

```python
reeses = cv2.imread('../Computer-Vision-with-Python/DATA/reeses_puffs.png',0)
```

```python
display(reeses, cmap='gray')
```

```python
cereals = cv2.imread('../Computer-Vision-with-Python/DATA/many_cereals.jpg', 0)
```

```python
display(cereals)
```

```python
orb = cv2.ORB_create()
```

```python
kp1,des1 = orb.detectAndCompute(reeses, None)
kp2,des2 = orb.detectAndCompute(cereals, None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
```

```python
matches = bf.match(des1,des2)
```

```python
matches = sorted(matches, key=lambda x:x.distance)
```

```python
reeses_matches = cv2.drawMatches(reeses, kp1, cereals,kp2,matches[:20],None, flags=2)
```

```python
display(reeses_matches)
```

```python
# using SIFT
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp1,des1 = sift.detectAndCompute(reeses, None)
kp2,des2 = sift.detectAndCompute(cereals, None)
```

```python
bf = cv2.BFMatcher()
```

```python
matches = bf.knnMatch(des1,des2,k=2)
```

```python
good = []

# less distance == better match
# ration match < 75% match2
for match1,match2 in matches:
    if match1.distance < 0.75*match2.distance:
        good.append([match1])
```

```python
good
```

```python
len(good)
```

```python
sift_matches = cv2.drawMatchesKnn(reeses, kp1, cereals, kp2, good, None, flags=2)
```

```python
display(sift_matches)
```

```python
# FLANN fast libraray for approx.nearest neighbours
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp1,des1 = sift.detectAndCompute(reeses, None)
kp2,des2 = sift.detectAndCompute(cereals, None)
```

```python
#FLANN
FLANN_INDEX_KDTREE = 0
index_params = dict(algorithm=FLANN_INDEX_KDTREE, trees=5)
search_params=dict(checks=50)
```

```python
flann = cv2.FlannBasedMatcher(index_params, search_params)
```

```python
matches = flann.knnMatch(des1,des2, k=2)
        
```

```python
good = []
for match1,match2 in matches:
    if match1.distance < 0.7*match2.distance:
        good.append([match1])
```

```python
flann_matches = cv2.drawMatchesKnn(reeses, kp1, cereals, kp2, good, None, flags= 2)
display(flann_matches)
```
