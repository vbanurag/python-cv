```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
full = cv2.imread('../Computer-Vision-with-Python/DATA/sammy.jpg')
full = cv2.cvtColor(full, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(full)
```

```python
face = cv2.imread('../Computer-Vision-with-Python/DATA/sammy_face.jpg')
face = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(face)
```

```python
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR','cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python
for m in methods:
    
    # Create a copy
    full_copy = full.copy()
    
    method = eval(m)
    
    # template matching
    
    res = cv2.matchTemplate(full_copy, face, method)
    
    min_val,max_val,min_loc,max_loc = cv2.minMaxLoc(res)
    
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
        
    height,width,channels = face.shape
    
    bottom_right = (top_left[0]+width, top_left[1]+height)
    
    cv2.rectangle(full_copy, top_left, bottom_right, (255,0,0), 10)
    
    # plot and show the images
    
    plt.subplot(121)
    plt.imshow(res)
    plt.title("Heat match of the template matching")
    
    plt.subplot(122)
    plt.imshow(full_copy)
    plt.title("Detection of template matching")
    # sub method
    plt.suptitle(m)
    
    # not override
    plt.show()
    
    print('\n')
    print('\n')
```
