### IN matplotlib channel read as RGB

### But in the Open CV chanel is reverse Blue Green Red (BGR)

```python
## open cv provide the function to change the order of the color
```

```python
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread('Computer-Vision-with-Python/DATA/00-puppy.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
plt.imshow(fix_img)
```

```python
new_img = cv2.resize(fix_img, (400, 400))
```

```python
plt.imshow(new_img)
```

```python
## fix image accordig to the ratio
```

```python
w_ratio=0.5
h_ratio=0.5
```

```python
new_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img, -1) ## for fliping the image and it will be good while creating the more dataset.
plt.imshow(new_img)   
```

```python
## saving the file
```

```python
cv2.imwrite('rotated_puppy.jpg', new_img)
```
